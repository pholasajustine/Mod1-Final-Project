```python
import json
from pandas import json_normalize
import pandas as pd
from datetime import date
```

Nous avons plusieurs fichiers dans lesquelles sont listées les offres d'alternances (provenant de l'api la bonne alternance). En effet, nous étions limitées à un rayon kilométrique de 200km. Nous avons donc dû en sortir plusieurs par rapport à différents points de la France pour essayer de couvrir la totalité du territoire.

```python
with open("Data/nord.json", "r", encoding="utf-8") as read_file:
    data_nord = json.load(read_file)
with open("Data/nord_est.json", "r", encoding="utf-8") as read_file:
    data_nord_est = json.load(read_file)
with open("Data/nord-ouest.json", "r", encoding="utf-8") as read_file:
    data_nord_ouest = json.load(read_file)
with open("Data/centre.json", "r", encoding="utf-8") as read_file:
    data_centre = json.load(read_file)
with open("Data/sud_est.json", "r", encoding="utf-8") as read_file:
    data_sud_est = json.load(read_file)
with open("Data/sud_ouest.json", "r", encoding="utf-8") as read_file:
    data_sud_ouest = json.load(read_file)
```

```python
data_centre.keys()
```

peJobs : les offres d'emploir provenant de pôle emploi

lbaCompanies : les offres d'emploi provenant de la bonna alternance

lbbCompanies : les offres d'emploi provenant de la bonne boîte

On utilise ici que les données provenant de peJobs car les autres source ne donne pas la date de création de l'offre, nous ne pourrions donc pas recouper ces données avec les données covid par date.

```python
df_list = []

#On concaténe les différents listes d'offres d'emploi provenant des différents fichiers

df_list.append(json_normalize(data_centre['peJobs']['results']))

df_list.append(json_normalize(data_nord_est['peJobs']['results']))

df_list.append(json_normalize(data_nord_ouest['peJobs']['results']))

df_list.append(json_normalize(data_sud_est['peJobs']['results']))

df_list.append(json_normalize(data_sud_ouest['peJobs']['results']))

df_list.append(json_normalize(data_nord['peJobs']['results']))

jobs_df = pd.concat(df_list, axis=0)

jobs_df
```

```python
jobs_df[['job.creationDate','title','place.city','company.name','place.latitude', 'place.longitude' ]]
```

Le fichier covid n'ayant pas la donnée par ville, on va ajouter les départements pour pouvoir faire le lien par département.

```python
def get_department(x):
    return((x[:2]))

jobs_df['place.departement'] = jobs_df['place.zipCode'].apply(str).apply(get_department)
```

Pour comparer avec le taux d'incidence covid par mois, on ajoute une colonne moi dans le dataframe

```python
def get_month(x):
    return(x[0:7])

jobs_df['month'] = jobs_df['job.creationDate'].apply(get_month)
```

```python
filtered_jobs_df = jobs_df[['job.creationDate','month','title','place.city','place.zipCode','place.departement','company.name']]

filtered_jobs_df
```

La date de création d'offres d'alternance la plus ancienne étant août 2020, il faut se restreindre dans notre fichier covid à prendre les données qu'à partir de ce mois-ci

```python
data_covid = pd.read_csv('Data/table-indicateurs-open-data-dep-2021-04-09-20h00.csv')

covid_df = pd.DataFrame(data_covid)
```

Pour pouvoir lier les deux fichiers entre eux, on ajoute une colonne mois au fichier covid.

```python
covid_df['month'] = covid_df['date'].apply(get_month)
```

Transformer les dates dans le bon format dans le but de pouvoir appliquer un filtre

```python
covid_df['date'] = pd.to_datetime(covid_df['date'], format='%Y-%m-%d')
```

```python
covid_df[['dep', 'date', 'month', 'tx_incid']]
```

Création d'un filtre pour n'obtenir les dates qu'à partir du 01-08-2021

```python
mask = (covid_df['date'] > '2020-08-1') & (covid_df['date'] <= '2021-04-11')
filtered_covid_df = covid_df.loc[mask]
data_covid_def = filtered_covid_df[['dep', 'date', 'month', 'tx_incid']]
data_covid_def
```

```python
final_data_covid_df = data_covid_def[data_covid_def.date.dt.day == 15]
final_data_covid_df
```

Exportation des 2 fichiers en .csv pour enregistrement dans une base de données SQL

```python
#Export 1ère source de données :

filtered_jobs_df.to_csv('Data/offres-alternance.csv')

#Export 2ème source de données :

final_data_covid_df.to_csv('Data/covid-incidence-dep.csv')
```
